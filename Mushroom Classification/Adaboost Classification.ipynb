```python
X = df.drop('class',axis=1)
```

```python
X
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
model = AdaBoostClassifier(n_estimators=1)  # Base Adaboost_Classifier
```

```python
model.fit(X_train,y_train)
```

```python
model.feature_importances_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
errors = []

for n in range(1,99):
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    error = 1 - accuracy_score(y_test,y_pred)
    errors.append(error)
```

```python
plt.plot(range(1,99),errors)
plt.xlabel('Number of estimators(no. of stumps)')
plt.ylabel('error rates')
```

```python
model = AdaBoostClassifier(n_estimators=99)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
feats = pd.DataFrame(data=model.feature_importances_,index=X.columns,columns=['Importance'])
```

```python
feats
```

```python
imp_feats = feats[feats['Importance'] > 0]
```
